## Import Libraries

```python
import pandas as pd
import numpy as np
import torch
from ucimlrepo import fetch_ucirepo
```

## Determine if GPU is enabled

```python
if torch.cuda.is_available():
    print("CUDA is available")
iris = fetch_ucirepo(id=53)
iris.data.features.head()
data = iris.data
data.features.head()
labels = data.targets
features = data.features
```

## Configure neural network and data

```python

model = torch.nn.Sequential(
    torch.nn.Linear(len(features.columns), 3, dtype=torch.float64),
    torch.nn.Sigmoid()
)
loss_fn = torch.nn.MSELoss(reduction="sum")
lr = 0.001
optimizer = torch.optim.Adam(model.parameters(), lr=lr)

one_hot = pd.get_dummies(labels).astype("float64")
one_hot.head()
numpy_one_hot = one_hot.to_numpy()
numpy_features = features.to_numpy()
feature_tensor = torch.from_numpy(numpy_features).double()
label_tensor = torch.from_numpy(numpy_one_hot).double()
print("Training data: ")
print(feature_tensor.dtype, label_tensor.dtype)
```

```python
for t in range(200000):
    y_pred = model(feature_tensor)

    loss = loss_fn(y_pred, label_tensor)
    if t%100 == 99:
        print(t, loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
```
